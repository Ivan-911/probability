```python
def mse_(B1, x = zp, y = ks, n = n):
    return np.sum((y - B1 * x)**2)/n
```

3. Произвести вычисления как в пункте 2, но с вычислением intercept. Учесть, что изменение коэффициентов должно производиться на каждом шаге одновременно (то есть изменение одного коэффициента не должно влиять на изменение другого во время одной итерации).

```python
def mse_(B1, B0, x = zp, y = ks, n = n):
    return np.sum((y -(B0 + B1 * x))**2)/n
```

```python
alpha = 0.0001
previous_mse = 1e+10 
threshold = 1e-5
i = 0
B1 = np.random.normal(loc=0, scale=1)
B0 = np.random.normal(loc=0, scale=1)
while 1:
#for i in range(3000):
    B1 -= alpha * np.sum((B0 + B1 * zp - ks) * zp) / len(zp)
    B0 -= alpha * np.sum(B0 + B1 * zp - ks) / len(zp)
    #gradient_B1 = np.sum((B0 + B1 * zp - ks) * zp) / len(zp)
    #gradient_B0 = np.sum(B0 + B1 * zp - ks) / len(zp)
    i += 1
    mse = mse_(B1, B0)
    previous_mse -= mse
    diff_mse = previous_mse
    if abs(previous_mse) < threshold:
        print('B1 = {B1}, B0 = {B0}, mse = {mse}, diff_mse={diff_mse}'.format(B1 = B1, B0=B0, mse=mse, diff_mse=diff_mse))
        break

    previous_mse = mse
    
    if i % 500 == 0:
        print('B1 = {B1}, mse = {mse}, B0 = {B0}, diff_mse={diff_mse}'.format(B1 = B1, B0 = B0, mse=mse, diff_mse=diff_mse))
        i = 0     
print('Итог B1: ', B1, 'Итог B0: ', B0, 'Итог mse: ', mse) 
```
